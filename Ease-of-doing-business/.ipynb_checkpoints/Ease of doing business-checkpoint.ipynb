```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_excel('rankings.xlsx')
```

```python
df.info()
```

```python
df.head()
```

```python
df.dropna(axis=0,inplace=True)
df.drop(['Rank within group'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(figsize=(15,15))
```

```python
print(df.columns)
```

```python
cols=['Starting a Business','Dealing with Construction Permits', 'Getting Electricity', 'Registering Property',
               'Getting Credit', 'Protecting Minority Investors', 'Paying Taxes', 'Trading across Borders',
               'Enforcing Contracts', 'Resolving Insolvency']

for item in cols:
    plt.scatter(df[item],df['globalRank'])
    plt.xlabel(item)
    plt.ylabel('Global Rank')
    plt.show()
```

```python
sns.pairplot(df,x_vars=['Starting a Business'],y_vars=['globalRank'])
```

```python
x=df[['Starting a Business','Dealing with Construction Permits', 'Getting Electricity','Registering Property',
           'Getting Credit','Protecting Minority Investors', 'Paying Taxes','Trading across Borders', 'Enforcing Contracts',
           'Resolving Insolvency']]

y=df[['globalRank']]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
x_train.hist(figsize=(15,15))
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
y_pred=lin_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r_sq=r2_score(y_pred,y_test)
print('R squared : {}'.format(r_sq))
rmse=mean_squared_error(y_pred,y_test)
print('RMSE : {}'.format(np.sqrt(rmse)))
mae=mean_absolute_error(y_pred,y_test)
print('MAE : {}'.format(mae))
```

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(lin_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
print(np.sqrt(-cvs))
print(np.sqrt(-cvs).mean())
print(np.sqrt(-cvs).std())
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg=RandomForestRegressor(n_estimators=10)
forest_reg.fit(x_train, y_train)
forest_pred=forest_reg.predict(x_test)
```

```python
r_sq=r2_score(forest_pred,y_test)
print('R squared : {}'.format(r_sq))
rmse=mean_squared_error(forest_pred,y_test)
print('RMSE : {}'.format(np.sqrt(rmse)))
mae=mean_absolute_error(forest_pred,y_test)
print('MAE : {}'.format(mae))
```

```python
cvs=cross_val_score(forest_reg,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
print(np.sqrt(-cvs))
print(np.sqrt(-cvs).mean())
print(np.sqrt(-cvs).std())
```

```python
print('Linear Regressor : {}'.format(lin_reg.predict(np.array([136,27,22,154,25,13,115,68,163,52]).reshape(1,10))))
print('Random Forest Regressor : {}'.format(forest_reg.predict(np.array([136,27,22,154,25,13,115,68,163,52]).reshape(1,10))))
```
